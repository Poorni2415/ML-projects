```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
A=pd.read_csv('/content/Global YouTube Statistics.csv',encoding='latin-1')
```

```python
A.head()
```

```python
A.columns
```

```python
B= A[['rank', 'Youtuber', 'subscribers', 'video views', 'category', 'Title',
       'uploads', 'Country', 'channel_type',
       'video_views_rank',
       'video_views_for_the_last_30_days']]
```

```python
B.head()
```

```python
pd.set_option("display.float_format","{:.2f}".format)
```

```python
B
```

```python
C=B.groupby("Country")["video views"].sum().sort_values(ascending=False)
```

```python
C
```

```python
B.groupby("category")["video views"].sum().sort_values(ascending=False)
```

```python
B.groupby("Country")["uploads"].sum().sort_values(ascending=False)
```

```python
C.plot(kind='bar', figsize=(10, 6))  # Use 'bar' for bar graph

plt.title('Total Video Views per Country')
plt.xlabel('Country')
plt.ylabel('Total Video Views')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```
