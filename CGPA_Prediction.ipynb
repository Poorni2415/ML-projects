```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score, r2_score, mean_squared_error
```

```python
a['1st']=a['1st'].fillna(a['1st'].mean() )
```

```python
a['2nd']= a['2nd'].fillna( a['2nd'].mean())
```

```python
a['3rd']=a['3rd'].fillna(a['3rd'].mean())
```

```python
a['4th']=a['4th'].fillna(a['4th'].mean())
```

```python
a['5th']=a['5th'].fillna( a['5th'].mean())
```

```python
a['Gender']=a['Gender'].fillna('Male')
```

```python
x=a.drop(columns='5th',axis=1)
y=a['5th']
```

```python
x
```

```python
y
```

```python
e=StandardScaler()
```

```python
A=e.fit_transform(x)
```

```python
A
```

```python
Y=pd.DataFrame(y)
B=e.fit_transform(Y)
```

```python
B
```

```python
A_train,A_test,B_train,B_test=train_test_split(A,B,random_state=1,test_size=0.1)
```

```python
from sklearn import svm
cl= svm.SVC(kernel='linear')
cl.fit(A_train,B_train)
```

```python
mod = LinearRegression()
mod.fit(A_train, B_train)
```

```python
tr=mod.predict(A_train)
tr_r2 = r2_score(B_train, tr)
tr_mse = mean_squared_error(B_train, tr)
tr_rmse = np.sqrt(tr_mse)
```

```python
print(f"R-squared: {tr_r2}")
print(f"MSE: {tr_mse}")
print(f"RMSE: {tr_rmse}")
```

```python
te=mod.predict(A_test)
te_r2 = r2_score(B_test, te)
te_mse = mean_squared_error(B_test, te)
te_rmse = np.sqrt(te_mse)
```

```python
print(f"R-squared: {te_r2}")
print(f"MSE: {te_mse}")
print(f"RMSE: {te_rmse}")
```
