```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
As=a.drop(columns='diagnosis',axis=1)
Bs=a['diagnosis']
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
A=sc.fit_transform(As)
B=Bs
```

```python
A
```

```python
B
```

```python
A.isnull().sum()
```

```python
A_train,A_test,B_train,B_test=train_test_split(A,B,test_size=0.1,random_state=0)
```

```python
mod=LogisticRegression()
```

```python
mod.fit(A_train,B_train)
```

```python
tr=mod.predict(A_train)
tr_acc=accuracy_score(tr,B_train)
```

```python
print(tr_acc)
```

```python
te=mod.predict(A_test)
te_acc=accuracy_score(te,B_test)
```

```python
print(te_acc)
```
